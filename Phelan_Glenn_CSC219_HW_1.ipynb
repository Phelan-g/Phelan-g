```python
#append that new DNA varible to the dna sequences-
dnaResult = [] #create an index to put the dna results into

for file in files:
  f= open(file)
  DNA = (f.readlines()[1:]) #index the file to start reading after the first line
  #print(DNA)
  dnaResult.append(DNA)
print(dnaResult[0])
print(dnaResult[1])
```

```python
#remove the newline character:
#the DNA lists that have printed so far include a '\n' at the end of every line DNA sequence (string)
#we want to print out the lists and strip the '\n at the eand of each string
tempList = [] #this will be a new index to store temporary values of a list in.
tempString = '' #this empty string will later be used in a for loop to store values of a string from a list
finalList = [] #this empty index will store the the value set of values of the file that will print the dna strings in a single list
for dnaList in dnaResult:
  #set up a for loop that will pull each dna list from the dnaResult index (list
  for itemString in dnaList:
    #grab each string inside each list
    sequence = itemString[:-1]  #the negative 1 will count the string from the end ('\n')
    tempList.append(sequence) #grab all string without '\n' and add it to a list
  tempString = ''.join(tempList)
  finalList.append(tempString)

print(finalList)
print(finalList[0]) #print the first item inthe list
print(finalList[1])#print the second item in the list
```

```python
from collections import Counter
#now write code to count the number of G and C variables in each of the sequewnces. 
#create an easily referenced data strucutre that contains the names of orgganisms from the fast file
#and GC variable count for each file. 
#Try using a dictionary for this

#counting G and C- 
#create a dictionary with the gene name and gene sequence {genename: geneseqence, genename:genesequence}
# then think about what we want to get back at the end, we want name and C or G counts, for each name,
# so this sounds like a dictionary.

genename = ['CCR5','cheetah','cat','malaria','polarbear']
genesequence = finalList[:]

genedict = {}

for i in range(len(genename)):
  genedict[genename[i]]= genesequence[i]

print(genedict)

#iterate through the dictionary
count=0 
resultDict = {}
    
for k,v in genedict.items(): #keys,val,  a,b
#the sequence is inthe v as a string and the gene name is in k
# so iterate over each character in the string
    for char in v:
      if char == 'C':
        count+= 1
      elif char == 'G':
        count += 1
    resultDict[k] = count 

print(resultDict)

print(char)#prints the last character of the sequence

```
